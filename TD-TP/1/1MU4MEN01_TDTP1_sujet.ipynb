**3.1** Le bloc d'instructions ci-dessous permet de tracer les isovaleurs de la fonction $f_1(x,y)$ pour $x \in [-4,4]$ et $y \in [-4,4]$. Exécuter le bloc d'instructions, puis analyser ce que fait chaque ligne. 

Faire varier les différents paramètres du tracé (intervalles de x et y, nombre de points de discrétisation, nombre d'isovaleurs) et analyser leur influence. En particulier, que se passe-t-il si la discrétisation du domaine est insuffisante ?

**3.3** Le bloc d'instructions ci-dessous donne le modèle de la fonction _df1_, qui calcule et renvoie le gradient de $f_1$. Compléter ce bloc d'instructions pour que _df1_ fasse bien ce qui est attendu. Utiliser ensuite cette fonction pour vérifier que le gradient de $f_1$ est bien nul aux points critiques.

```python
def df1 (x,y) :
    # Dérivée partielle par rapport à x
    df_dx = 3*x**2 - 3
    # Dérivée partielle par rapport à y
    df_dy = 3*y**2 - 12
    return [df_dx,df_dy]
```

```python
# A compléter pour calculer le gradient de f1 aux points critiques
def Grad_F1(x,y) : 
    Grad = df1(x,y)
    if Grad == [0,0] : 
        print("Gradient nul pour le couple de points", x , ";", y)

    
Grad_F1(-1,-2) ;  Grad_F1(-1,2) ; Grad_F1(1,-2) ; Grad_F1(1,2) 
```

**3.4** Le bloc d'instructions ci-dessous donne le modèle de la fonction _d2f1_, qui calcule et renvoie la matrice hessienne de $f_1$. Compléter ce bloc d'instructions pour que _d2f1_ fasse bien ce qui est attendu. Utiliser ensuite cette fonction pour vérifier la nature des points critiques de $f_1$.

```python
# Bloc d'instructions à compléter
def d2f1 (x,y) :
   d2f_dxdx =  6*x
   d2f_dxdy =  0
   d2f_dydx =  0
   d2f_dydy =  6*y
         
   H = [[d2f_dxdx,d2f_dxdy],[d2f_dydx,d2f_dydy]]
   # Dérivées partielles d'ordre 2
   return H
      
#Exemple pour (-1,2)
d2f1(-1,2)
```

```python
# A compléter pour déterminer la nature des points critiques de f1


# Calcul de la matrice hessienne

PointsCrit = [ [-1 , 2] , [-1 ,-2] , [1 ,2] , [1 ,-2] ] 
for i in  range(0,4,1): # de 0 à 3
    x,y = PointsCrit[i] # i eme ligne Matricielle 
    Hf1 = d2f1(x,y) # Matrice Hessienne
    ValeursPropres = np.linalg.eigvals(Hf1)
    print('Valeurs propres du couple', x ,y ,'de la matrice {} :'.format(Hf1))
    print('  vp1 = {}'.format(ValeursPropres[0]))
    print('  vp2 = {}'.format(ValeursPropres[1]))
    
    if  ValeursPropres[0] > 0 and ValeursPropres[1]  > 0 : 
        print("Le point critique est un Min local","\n")
    
    if  ValeursPropres[0] < 0 and ValeursPropres[1]  < 0 : 
        print("Le point critique est un Max local","\n")
    
    if  ValeursPropres[0]== 0 or ValeursPropres[1] ==  0 : 
        print("Pas de conclusion possible","\n")
    
    if np.sign(ValeursPropres[0]) != np.sign(ValeursPropres[1]) : 
         print("Le point critique est un col","\n") # n == saut de ligne



    """"Note: Il est inutile de calculer les valeurs propres dans ce cas
     puisque la matrice hessienne est deja diagonale dans la base formée
     par ses Vecteurs propres. Le signe de la valeure propre sera 
     Conditionnée par le signe de x et y """
   
    
# Calcul des valeurs propres de la matrice hessienne, puis test de leur signe
```

```python
# Matrice Hessienne 

def d2f2 (x,y) :
   d2f2_dxdx =  6*x
   d2f2_dxdy =  -1
   d2f2_dydx =  -1
   d2f2_dydy =  48*y
         
   H2 = [[d2f2_dxdx,d2f2_dxdy],[d2f2_dydx,d2f2_dydy]]
   # Dérivées partielles d'ordre 2
   return H2
      
#Exemple pour (0,0)
d2f2(0,0)
```

```python
# Points critiques 


PointsCrit2 = [ [2 , 1] , [0 , 0] ] 
L = len(PointsCrit2)
for i in range(L) :
    x,y = PointsCrit2[i] # i eme ligne Matricielle 
    H2 = d2f2(x,y) # Matrice Hessienne pour chaque couple de points critiques
    ValeursPropres = np.linalg.eigvals(H2)
    print('Valeurs propres du couple', x ,";",y ,'de la matrice {} :'.format(H2))
    print('  vp1 = {}'.format(ValeursPropres[0]))
    print('  vp2 = {}'.format(ValeursPropres[1]))
    
    if  ValeursPropres[0] > 0 and ValeursPropres[1]  > 0 : 
        print("Le point critique est un Min local","\n")
    
    if  ValeursPropres[0] < 0 and ValeursPropres[1]  < 0 : 
        print("Le point critique est un Max local","\n")
    
    if  ValeursPropres[0]== 0 or ValeursPropres[1] ==  0 : 
        print("Pas de conclusion possible","\n")
    
    if np.sign(ValeursPropres[0]) != np.sign(ValeursPropres[1]) : 
         print("Le point critique est un col","\n") # n == saut de ligne
```

```python
# Matrice Hessienne 

def d2f3 (x,y) :
   d2f3_dxdx =  2 + 120*x**2 - 40*y
   d2f3_dxdy =  -40*x
   d2f3_dydx =  -40*x
   d2f3_dydy =  20
         
   H3 = [[d2f3_dxdx,d2f3_dxdy],[d2f3_dydx,d2f3_dydy]]
   # Dérivées partielles d'ordre 2
   return H3
      
#Exemple pour (0,0)
d2f3(1,1)
```

```python
# Points critiques 


PointsCrit3 =  [1 , 1] 

x,y = PointsCrit3 # i eme ligne Matricielle 
H3 = d2f3(x,y) # Matrice Hessienne pour chaque couple de points critiques
ValeursPropres = np.linalg.eigvals(H3)
print('Valeurs propres du couple', x ,";",y ,'de la matrice {} :'.format(H3))
print('  vp1 = {}'.format(ValeursPropres[0]))
print('  vp2 = {}'.format(ValeursPropres[1]))
    
if  ValeursPropres[0] > 0 and ValeursPropres[1]  > 0 : 
        print("Le point critique est un Min local","\n")
    
if  ValeursPropres[0] < 0 and ValeursPropres[1]  < 0 : 
        print("Le point critique est un Max local","\n")
    
if  ValeursPropres[0]== 0 or ValeursPropres[1] ==  0 : 
        print("Pas de conclusion possible","\n")
    
if np.sign(ValeursPropres[0]) != np.sign(ValeursPropres[1]) : 
         print("Le point critique est un col","\n") # n == saut de ligne
```
